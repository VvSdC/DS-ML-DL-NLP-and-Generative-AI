#### Data Cleaning

##### Reviews

```python
# Number of rows in dataset is 10841
# But, the number of reviews that are numeric are 10840
#There is one non-numeric value for reviews
df['Reviews'].str.isnumeric().sum()
```

```python
# Finding the row that has a non-numeric review value
df[~df['Reviews'].str.isnumeric()]
```

```python
# Dropping the row with non numeric review data
df_copy = df.copy()
df_copy = df_copy.drop(df_copy.index[10472])
df_copy.head()
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
# Convert reviews column into integer
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
df_copy.info()
```

##### Size

```python
df_copy['Size'].unique()
```

```python
# Converting all into K units

df_copy['Size'] = df_copy['Size'].str.replace('M','000')
df_copy['Size'] = df_copy['Size'].str.replace('k','')
df_copy['Size'].unique()
```

```python
# The last row has Size as 'Varies with device'
df_copy[~df_copy['Size'].str.isalnum()]
```

```python
# Replacing varies with device with NaN

df_copy['Size'] = df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size'] = df_copy['Size'].astype(float)
df_copy.info()
```

##### Cleaning Installs and Price

```python
df_copy['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
chars_to_remove = ['+',',','$']
cols_to_clean = ['Installs','Price']

for item in chars_to_remove:
    for col in cols_to_clean:
        df_copy[col] = df_copy[col].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
df_copy['Price']=df_copy['Price'].astype(float)
df_copy.info()
```

##### Last updated feature

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy.head()
```

```python
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
df_copy.head()
```

```python
# We can drop last updated
df_copy.drop('Last Updated',axis=1,inplace=True)
df_copy.head()
```

```python
df_copy.info()
```

### Exploratory Data Analysis

```python
# Finding the number of duplicate rows
df_copy[df_copy.duplicated('App')].shape
```

```python
# Dropping duplicates
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_copy.shape
```

#### Exploring data

```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != "str"]
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == "str"]

print('We have {} numerical features: {}'.format(len(numeric_features),numeric_features))
print('We have {} categorical features: {}'.format(len(categorical_features),categorical_features))
```

```python
# Proportions of count data on categorical columns

for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('-'*20)
```

```python
# Proportions of count data on numerical columns

plt.figure(figsize=(15,15))
plt.suptitle("Univariate Analysis of Numerical Features",fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0,len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x = df_copy[numeric_features[i]], shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

### Observations from numerical columns plots

RIGHT SKEWED: Reviews, Size, Installs and Price

LEFT SKEWED: Rating and Year

```python
# For categorical columns

plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis of Categorical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
category=['Type','Content Rating']

for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

### Which is the most popular app category?

```python
df_copy['Category'].value_counts().plot.pie(y=df['Category'],figsize=(15,6),autopct='%1.1f')
```

#### Top 10 App Categories

```python
category = pd.DataFrame(df_copy['Category'].value_counts())
category.rename(columns={'Category':'Count'},inplace=True)
category
```

```python
# Creating bar chart for top 10 categories

plt.figure(figsize=(15,6))

category_reset = category.reset_index()
category_reset.columns = ['Category','Count']

sns.barplot(x='Category', y='Count', data=category_reset[:10], palette='hls')
plt.title("Top 10 App categories")
plt.xticks(rotation=90)
plt.show()
```

### Which category has largest number of installations?

```python
category_wise_installations = df_copy.groupby('Category')['Installs'].sum().sort_values()

plt.figure(figsize=(15,6))
sns.barplot(category_wise_installations[:5])
```

```python
# get top-10 by installs (ensure descending)
top10 = category_wise_installations.sort_values(ascending=False).head(10)

plt.figure(figsize=(8,8))
colors = sns.color_palette('tab10', n_colors=len(top10))
plt.pie(top10.values,
        labels=top10.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=colors,
        textprops={'fontsize':9})
plt.title('Top 10 Categories by Installations')
plt.axis('equal')
plt.show()
```
