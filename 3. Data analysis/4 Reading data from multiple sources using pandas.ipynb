### JSON

```python
from io import StringIO
import pandas as pd

# JSON data with two records (list of dicts)
json_data = '''
[
  {"id": 101, "name": "Alice", "department": "Engineering",
   "skills": ["Python", "Pandas", "SQL", "Machine Learning"]},
  {"id": 102, "name": "Bob", "department": "Marketing",
   "skills": ["Excel", "SEO", "Content Writing"]}
]
'''

# Convert string into a file-like object
string_data = StringIO(json_data)

# Read into pandas DataFrame
df = pd.read_json(string_data)
print(df)
```

```python
# Converting back to json
# It adds index to each value for a column

df.to_json()
```

```python
# Using orient as index
# We will get each row as a value for index

df.to_json(orient='index')
```

```python
# Using orient as records
# We will get each row as an element in list

df.to_json(orient='records')
```
