## Matplotlib

Matplotlib is a powerful plotting library for python that enables the creation of static, animated and interactive visualizations. It is widely used for data visualization in data science and analytics.

```python
import matplotlib.pyplot as plt
```

#### Basic plots

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Plot with x list in x-axis and y in y-axis
plt.plot(x,y)

# Labelling X-axis and Y-axis in the plot
plt.xlabel('Natural Numbers')
plt.ylabel('Squares of natural numbers')

# Providing title for the plot regarding what plot is about
plt.title("Curve of natural numbers and their corresponding squares")

# Display plot
plt.show()
```

#### Customized plots

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Supported values of linestyle are ['-','--','-.',':','None',' ','','solid','dashed','dotted','dashdot']
plt.plot(x,y,color='green',linestyle='-.')
```

```python
# Adding a marker
# Marker adds a symbol at each point (x,y)

plt.plot(x,y,color='red',linestyle='--',marker='o')
```

```python
# Making lines thicker

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3)
```

```python
# Increasing marker size and adding a grid

plt.plot(x,y,color='red',marker='o',linestyle='--',linewidth=3,markersize=9)
plt.grid(True)
```

#### Multiple plots

```python
# Import math
import math

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,8,27,64,125]
y3 = [math.log(1),math.log(2),math.log(3),math.log(4),math.log(5)]
y4 = [math.sin(1),math.sin(2),math.sin(3),math.sin(4),math.sin(5)]

plt.figure(figsize=(13,6))

# First plot
# The first plot in 2 row x 2 column plots
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Natural numbers and their corresponding squares")


# Second plot
plt.subplot(2,2,2)
plt.plot(x,y2,color='red')
plt.title("Natural numbers and their cubes")


# Third plot
plt.subplot(2,2,3)
plt.plot(x,y3,color='blue')
plt.title("Natural numbers and their natural logarithms")


# Fourth plot
plt.subplot(2,2,4)
plt.plot(x,y4,color='yellow')
plt.title("Natural numbers and their sin values")
```

#### Bar plots

```python
role = ['SP_L1','SP_L2','SP_L3','DSE']
salary = [10,16,21,7.50]

plt.bar(role,salary,color='purple')
plt.xlabel("Infosys hiring roles")
plt.ylabel("Annual salary")
plt.title("Details about infosys hiring")
plt.show()
```

#### Histogram

Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

```python
data = [1,1,2,2,2,2,2,2,3,3,3,4,4,5,5,5]

plt.hist(data,bins=5,color='red',edgecolor='blue')
```

### Scatter plot

```python
x = [1,2,3,4,5]
y = [0,1,4,9,16]

plt.scatter(x,y,color='orange',marker='x')
```

### Pie chart

```python
labels = ['Fixed deposits','Mutual funds','Gold/Silver','Expenses','Bills']
sizes = [20,20,10,35,15]
colors = ['green','cyan','gold','red','orange']
explode = [0,0,0,0.1,0.05]

plt.pie(sizes,labels=labels,colors=colors,explode=explode)
```

```python
# Adding percentages to plot
# Using autopct will show percentage of each section in the plot

plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%")
```
