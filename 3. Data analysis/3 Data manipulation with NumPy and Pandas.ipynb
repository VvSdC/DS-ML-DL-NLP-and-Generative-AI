## Data Manipulation and Analysis with Pandas

Data manipulation and analysis are key tasks in any data science or data analysis project. Pandas provides a wide range of functions for data manipulation and analysis, making it easier to clean, transform and extract insights from data.

```python
import pandas as pd
```

```python
df = pd.read_csv('employees.csv')

print(f"First 5 rows of the dataset are: \n{df.head(5)}")
print(f"\nLast 5 rows of the dataset are : \n{df.tail(5)}")
```

```python
# Finding missing values in the dataset

print(f"Checking whether each column has at least one null value or not: \n{df.isnull().any()}")
print(f"\nNumber of null values in each column is : \n{df.isnull().sum()}")
```

```python
# Filling all null values with 0

df_filled_zeros = df.fillna(0)
df_filled_zeros
```

```python
# Filling all null values with the mean of that column
df_filled_with_mean = df

df_filled_with_mean['salary'] = df['salary'].fillna(df['salary'].mean())
df_filled_with_mean['age'] = df['age'].fillna(df['age'].mean())

df_filled_with_mean
```

```python
# Renaming  columns

df = df.rename(columns={'salary':'inhand_salary'})
df.head()
```

```python
# Converting data type of a column

df['int_salary'] = df['inhand_salary'].astype(int)
df.head()
```

```python
# Finding yearly salary

df['annual_salary'] = df['inhand_salary'].apply(lambda x : x * 12)
df.head()
```

### Data aggregating and grouping

```python
grouped_mean = df.groupby('department')['annual_salary'].mean()
print(f"Mean annual salary for each department is : \n{grouped_mean}")
```

```python
# Multiple column grouping

grouped_mean = df.groupby(['department','name'])['inhand_salary'].mean()
grouped_mean
```

```python
# Multiple aggregators

grouped_aggregation = df.groupby('department')['inhand_salary'].agg(['mean','sum','count'])
print(f"Different inhand_salary aggregation values after grouping by department is : \n{grouped_aggregation}")
```

### Merging and joining data

```python
df1 = pd.DataFrame({'key':['A','B','C','D'],'value1':[1,2,3,4]})
df2 = pd.DataFrame({'key':['A','C','D','E'],'value2':[9,10,11,12]})

print(f"DataFrame df1 is : \n{df1}")
print(f"\nDataFrame df2 is : \n{df2}")
```

```python
# Merge df using inner join using key
# Since we use inner join, we will only get common rows where keys match

pd.merge(df1,df2,on='key',how='inner')
```

```python
# Merging dataframes using outer join using key
# Since we are performing outer join, we will get both common and unique elements of both dataframes

pd.merge(df1,df2,on='key',how='outer')
```

```python
# Merging both dataframes using left outer join on 'key'
# Since we are performing left_outer join, we will get common key values of both dataframes and unique values of left dataframe

pd.merge(df1,df2,on='key',how='left')
```

```python
# Merging both dataframes using right outer join on 'key'
# We will get common 'key' fields from both dataframes and unique values of right dataframe

pd.merge(df1,df2,on='key',how='right')
```
