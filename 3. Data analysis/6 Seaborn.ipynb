```python
import seaborn as sns
```

### Basic plotting with seaborn

We can load some default datasets present in seaborn and then plot some graphs based on it.

```python
tips_dataset = sns.load_dataset('tips')
tips_dataset
```

```python
# Creating a scatter plot
import matplotlib.pyplot as plt 

sns.scatterplot(x='total_bill',y='tip',data = tips_dataset)
plt.title("Scatter plot of total bill vs tips")
```

```python
# Line plot

sns.lineplot(x = 'size', y = 'total_bill' , data = tips_dataset)
plt.title("Line plot of total bill by size")
```

```python
# Categorical plots

sns.barplot(x = 'day', y = 'total_bill', data = tips_dataset)
plt.title("Bar plot of total bill by day")
```

```python
# Box plot

sns.boxplot(x = 'day', y = 'total_bill' , data= tips_dataset)
plt.title("Box plot of total_bill by day")
```

```python
# Violin plot

sns.violinplot(x = 'day', y = 'total_bill', data= tips_dataset)
plt.title("Violin plot between total_bill by day")
```

```python
# Histogram
# kde -> Kernel density estimate (smoothens the histogram)

sns.histplot(tips_dataset['total_bill'],bins= 10,kde=True)
plt.title("Histogram to total_bills")
```

```python
# KDE plot
# fill = False -> Just gives the smoothened curve and no color within its region

sns.kdeplot(tips_dataset['total_bill'],fill=True)
plt.title("Kernel Density Estimate plot of total_bills")
```

```python
# Pairplot
# Used to find relationship between variables

sns.pairplot(tips_dataset)
```

```python
# Heatmap
#Finding correlation
corr = tips_dataset[['total_bill','tip','size']].corr()
print(f"The correlation between variables is : \n{corr}\n")

sns.heatmap(corr,annot=True,cmap='coolwarm')
```
