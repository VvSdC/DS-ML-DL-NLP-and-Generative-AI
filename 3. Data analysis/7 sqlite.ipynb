## SQL and SQLite

SQL is a standard language for managing and manipulating relational databases.
SQLite is a self-contained, serverless and zero-configuration database engine that is widely used for embedded database systems.

```python
import sqlite3
```

```python
# Connect to an SQLite database

connection = sqlite3.connect('learning.db')
connection
```

```python
# Creating cursor that iterates through all tables

cursor = connection.cursor()
```

```python
# Creating a table

# Writing a query
cursor.execute('''
    CREATE TABLE IF NOT EXISTS EMPLOYEES (
        id INTEGER PRIMARY KEY,
        name Text NOT NULL,
        age INTEGER,
        department TEXT           
    )
''')


# Commit the changes
# When we are modifying db by either inserting/updating or deleting data, we must perform commit to make those changes persist in the database
connection.commit()
```

```python
# Inserting data into a table

cursor.execute('''
    INSERT INTO EMPLOYEES (name, age, department) VALUES
    ('Alice Johnson', 28, 'HR'),
    ('Bob Smith', 34, 'Finance'),
    ('Charlie Davis', 41, 'IT'),
    ('Diana Lee', 25, 'Marketing'),
    ('Ethan Brown', 30, 'Sales'),
    ('Fiona White', 38, 'Operations'),
    ('George Miller', 29, 'Finance'),
    ('Hannah Wilson', 33, 'IT'),
    ('Ian Clark', 27, 'HR'),
    ('Julia Adams', 36, 'Marketing')
''')

connection.commit()
```

```python
# Obtaining data from a SQLite3 table
cursor.execute('''
    SELECT * FROM EMPLOYEES
''')

# Obtaining all rows from table
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
# Update data in the table
cursor.execute('''
    UPDATE EMPLOYEES SET AGE = 40 WHERE NAME = 'Ian Clark'
''')
connection.commit()

# Confirming changes
cursor.execute('''
        SELECT * FROM EMPLOYEES WHERE NAME = 'Ian Clark'
''')
row = cursor.fetchone()
print(row)
```

```python
# Delete data from the table
cursor.execute('''
    DELETE FROM EMPLOYEES WHERE NAME = 'Bob Smith'
''')
connection.commit()

#Confirming changes
cursor.execute('''
    SELECT * FROM EMPLOYEES
''')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

### Inserting multiple rows of data into SQLite3 table

```python
# Creating a table for sales
cursor.execute('''
    CREATE TABLE SALES (
        date TEXT,
        product TEXT,
        total_sales INTEGER,
        region TEXT
    )
''')
connection.commit()


sales_data = [
    ('2026-01-05', 'Laptop', 15, 'North'),
    ('2026-01-06', 'Smartphone', 25, 'East'),
    ('2026-01-07', 'Tablet', 10, 'West'),
    ('2026-01-08', 'Headphones', 40, 'South'),
    ('2026-01-09', 'Smartwatch', 18, 'North'),
    ('2026-01-10', 'Monitor', 12, 'East'),
    ('2026-01-11', 'Keyboard', 30, 'West'),
    ('2026-01-12', 'Mouse', 35, 'South'),
    ('2026-01-13', 'Printer', 8, 'North'),
    ('2026-01-14', 'Camera', 20, 'East')
]

cursor.executemany('''
    INSERT INTO SALES(date,product,total_sales,region) VALUES (?,?,?,?)
''',sales_data)

connection.commit()

# Confirming changes
cursor.execute('''
    SELECT * FROM SALES
''')
rows = cursor.fetchall()
for row in rows:
    print(row)
```
