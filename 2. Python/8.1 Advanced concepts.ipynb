### Iterators

Iterators are advanced python concepts that allow for efficient looping and memory management.Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure

```python
# Not lazy loaded

my_list = [1,2,3,4,5]
for element in my_list:
    print(element,end=' ')
```

```python
# Iterators are lazy loaded

iterator = iter(my_list)
print(type(iterator))
print(iterator)
```

```python
# Loading only the first element  
# Using next function will take us to next consecutive elements in the list

next(iterator)
```

```python
# To iterate the entire list by lazy loading each element

new_iterator = iter(my_list)

while True:
    try:
        element = next(new_iterator)  
        print(element,end=' ')           
    except StopIteration:
        print("\nAll elements in the list are iterated")
        break
```

### Generators

Generators are simpler way to create iterators. It is a subset of iterators. They use the yield keyword to produce a series of values lazily, which means they generate a series of values on the fly and donot store them in memory.

So, generators are particularly used for reading large files because they allow us to process one line at a time without loading the entire file into the memory.

```python
# Ideally since first element square itself is returned, we get 0 since it is the first index and it's square is zero

def square(n):
    for i in range(n):
        return i**2
    
square(4)
```

```python
# But when generator is used, we get all the elements even if only the first one is returned

def square(n):
    for i in range(n):
        yield i**2

print(square(4))

# The values are generated on a fly and not stored somewhere in memory
for element in square(4):
    print(element,end=' ')
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
print(gen)

while True:
    try:
        element = next(gen)
        print(element)
    except StopIteration:
        print("All elements are obtained")
        break
```

```python
# Reading large files

def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line


file_content = read_large_file('text.txt')

while True:
    try:
        print(next(file_content).strip())
    except StopIteration:
        print("Entire file printed")
        break
```

### Function copy, closures and decorators

##### Closure
Closures are the functions that exists inside another function

```python
# Function copy

def welcome():
    return "This is the original welcome function"

# Assigning function to a variable
wel = welcome

# Deleting the actual function
del welcome

# Even after deleting the function, the copy that we have created still exists
print(wel())
```

```python
# Closures

def main_welcome():
    msg = "Welcome"
    def sub_welcome():
        print("This is the sub welcome function")
        print(f"The variable present inside the method can be accessed by sub method: {msg}")
        print("This closures topic is interesting")
    return sub_welcome()

main_welcome()
```

```python
# Giving in-built function in closure

def main_welcome(func):
    msg = "I am doing well!"
    def sub_welcome():
        print("I am inside sub_welcome function")
        print(f"The length of msg string declared in main_welcome method is {func(msg)}")
    print("I am outside sub_welcome and inside main_welcome")
    return sub_welcome()


# Passing len inbuilt method which calculates length of the string
main_welcome(len)
```

#### Decorators
Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code.

```python
# Normal closure

def main_welcome(func):
    msg = "This is a variable from main_welcome method"
    def sub_welcome():
        print("I am inside sub_welcome method")
        print(f"Variable msg accessed from outside sub_welcome method has value : {msg}")
        func()
    print("This is inside main_welcome method")
    return sub_welcome()

def outside_welcome():
    print("This is from outside welcome function")

main_welcome(outside_welcome)
```

```python
# Using decorator

def main_welcome(func):
    msg = "This is a variable from main_welcome method"
    def sub_welcome():
        print("I am inside sub_welcome method")
        print(f"Variable msg accessed from outside sub_welcome method has value : {msg}")
        func()
    print("This is inside main_welcome method")
    return sub_welcome()

# This will directly send the outside_welcome method as parameter to main_welcome method
@main_welcome
def outside_welcome():
    print("This is from outside welcome function")
```

```python
# Decorators with arguments

def repeat(n):
    def decorator(func):
        def wrapper():
            for _ in range(n):
                func()
        return wrapper
    return decorator


@repeat(3)
def print_name():
    print("Charan")

print_name()
```
