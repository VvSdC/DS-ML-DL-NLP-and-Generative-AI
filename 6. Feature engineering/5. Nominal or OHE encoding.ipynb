### Data encoding

Data encoding is a technique to convert categorical data into numerical values so that machines can understand them.
1. Nominal/OHE encoding
2. Label and ordinal encoding
3. Target guided ordinal encoding

### Nominal/OHE encoding

One hot encoding, also known as nominal encoding is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green and blue), we can represent it in one hot encoding as follows:

1. Red: [1,0,0]
2. Green [0,1,0]
3. Blue [0,0,1]

#### Drawback of One Hot Encoding / Nominal Encoding

One‑Hot Encoding transforms categorical variables into binary vectors. While effective for small sets of categories, it has a significant limitation:

- High Dimensionality: When a categorical feature contains a large number of unique values (high cardinality), One‑Hot Encoding produces a very large, sparse matrix.
- Sparsity Issues: Most entries in the encoded matrix are zeros, which increases memory usage and computational overhead.
- Risk of Overfitting: Machine learning models may overfit to noise in such high‑dimensional data, reducing generalization performance.
- Reduced Efficiency: Training and inference become slower due to the inflated feature space.

Because of these drawbacks, One‑Hot Encoding is generally not recommended for categorical variables with many unique values.

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color': ['red','blue','green','red','blue']
})
df.head()
```

```python
# Create an instance of OneHotEncoder

encoder = OneHotEncoder()
```

```python
encoded_data = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out())
encoded_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoded_df],axis=1)
```
