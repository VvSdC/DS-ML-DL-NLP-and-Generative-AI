### Handling imbalanced dataset

If there is a dataset with majority of data points belonging to a single class then using that data to train can make model biased towards majority class i.e if a data set contains 1000 data points with 900 belonging to class A and 100 belonging to class B, then the model will be biased towards labeling data points to class A

```python
import numpy as np 
import pandas as pd 
```

```python
# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with 2 classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
# Creating dataframe with imbalanced dataset

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1]*n_class_1
})
```

```python
df = pd.concat([class_1,class_0]).reset_index(drop=True)
df.head()
```

```python
df['target'].value_counts()
```

#### Upsampling

To reduce the imbalance in the dataset, we increase the number of 1's (in this case) to get a balance between number of data points with class 1 and 0.

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
from sklearn.utils import resample

df_minority_upsampled = resample(df_minority, replace=True, n_samples=len(df_majority), random_state=42)
df_minority_upsampled
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upsampled])
df_upsampled['target'].value_counts()
```

### Down sampling

In down sampling we will reduce the number of majority samples (number of data points with target 0 in this case)

```python
# Creating dataframe with imbalanced dataset

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1]*n_class_1
})
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
df['target'].value_counts()
```

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
df_majority_downsampled = resample(
    df_majority, replace=False,
    n_samples=len(df_minority),
    random_state=42
)
df_majority_downsampled.shape
```

```python
df_downsampled = pd.concat([df_majority,df_minority_upsampled])
df_downsampled['target'].value_counts()
```
