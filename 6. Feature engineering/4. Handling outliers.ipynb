### 5 Number summary and box plot

Minimum, Maximum, Median, Median, Q1, Q3, IQR

```python
import numpy as np
```

```python
marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum,Q1,median,Q3,maximum = np.quantile(marks,[0,0.25,0.50,0.75,1.0])
```

```python
print(f"Minimum: {minimum}")
print(f"Q1: {Q1}")
print(f"Median: {median}")
print(f"Q3: {Q3}")
print(f"Maximum: {maximum}")
```

```python
IQR = Q3 - Q1
print(f"IQR is : {IQR}")
```

```python
lower_fence = Q1 - 1.5*IQR
higher_fence = Q3 + 1.5*IQR

print(f"Lower fence is: {lower_fence}")
print(f"Higher fence is: {higher_fence}")
```

```python
import seaborn as sns
```

```python
sns.boxplot(marks)
```

```python
marks_with_outliers = marks + [-100.5,123,145,186,-5]
sns.boxplot(marks_with_outliers)
```
