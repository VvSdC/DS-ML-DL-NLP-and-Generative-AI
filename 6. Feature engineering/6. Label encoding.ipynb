### Label encoding

Label encoding and ordinal encoding are two techniques used to encode categorical data as numerical data.

Label encoding involves assigning a unique numerical label to each category in the variable. The labels are usually assigned in alphabetical order or based on the frequency of the categories. For example, if we have a categorical variable "color" with three different values (red, green, blue), we can represent it using label encoding as follows:

1. Red: 1
2. Green: 2
3. Blue: 3

### When to use label encoding

Label encoding can be used when categorical values are not nominal i.e in case of colors etc where categories cannot be ranked. For ordinal values like high,low etc where there are ranks, we use ordinal encoding

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
label_encoder = LabelEncoder()
```

```python
colors_df = pd.DataFrame({
    'color': ['red','blue','green','red','blue']
})
label_encoder.fit_transform(colors_df[['color']])
```

```python
label_encoder.transform([['red']])
```

### When to use ordinal encoding

Ordinal encoding can be used when categorical values have an order like high,low etc where rank matters. Another example can be qualification like:

Post Graduate > Graduate > College > High School

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'sizes': ['small','medium','large','medium','small','large']
})
df.head()
```

```python
## Assigning ranks
ordinal_encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
ordinal_encoder.fit_transform(df[['sizes']])
```

```python
ordinal_encoder.transform([['medium']])
```
